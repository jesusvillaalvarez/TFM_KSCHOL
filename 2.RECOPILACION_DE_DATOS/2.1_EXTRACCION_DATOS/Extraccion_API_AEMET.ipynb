## WEBs de calidad del AIRE


La calidad del aire y el nivel de partículas en suspensión son medidas comunes que se proporcionan a través de diversas APIs. Algunas fuentes confiables y populares para obtener datos sobre la calidad del aire incluyen:



1.   **World Air Quality Index (WAQI):**
Sitio web: World Air Quality Index
API: WAQI proporciona una API que puedes utilizar para obtener datos sobre la calidad del aire en todo el mundo.
Endpoint de la API: https://api.waqi.info/feed/geo:{lat};{lon}/?token=YOUR_API_KEY
Necesitarás registrarte para obtener una clave de API.

2.   **OpenAQ:**
Sitio web: OpenAQ
API: OpenAQ ofrece una API que permite acceder a datos globales de calidad del aire.
Endpoint de la API: https://api.openaq.org/v1/measurements
Puedes consultar la documentación para más detalles y opciones.
AirNow API (para EE. UU.):

3.   **AirNow:**
Sitio web: AirNow
API: AirNow proporciona una API que ofrece información sobre la calidad del aire en los Estados Unidos.
Endpoint de la API: https://www.airnowapi.org/aq/observation/latLong/current/?format=application/json&latitude={lat}&longitude={lon}&distance=25&API_KEY=YOUR_API_KEY
Necesitarás registrarte y obtener una clave de API.

3.   **Plume Labs:**
Plume Labs - Plume API:
Sitio web: Plume Labs

Endpoint de la API: Puedes obtener más información sobre la API al ponerte en contacto con Plume Labs.
